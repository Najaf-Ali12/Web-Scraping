```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
descriptions=[]
processor_data=[]
ram_data=[]
windows_data=[]
storage_data=[]
display_data=[]
price_data=[]
rating=[]
pages=list(range(1,100))
for page in pages:
    url="https://www.flipkart.com/search?q=Laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}".format(page)
    r=requests.get(url)
    soup= BeautifulSoup(r.text,"lxml")
    #print(soup.prettify())  # changes in prettily looking human readable string
    containers=soup.find_all("div",class_="ZFwe0M row")  # total 24 laptops shown on one page
    if len(containers)==0:
        print("Kuch tu gadbad hae daya")
    else:
        print(page)
    for container in containers:
        rate=container.find("div",class_="MKiFS6")
        rating.append(rate.get_text() if rate else "N/A")

        description=container.find("div",class_="RG5Slk")
        descriptions.append(description.get_text(strip=True) if description else "N/A")

        processor=container.find("li",class_="DTBslk")
        processor_data.append(processor.get_text(strip=True) if processor else "N/A")

        ram=container.find("li",class_="DTBslk").find_next_sibling("li")
        ram_data.append(ram.get_text(strip=True) if ram else "N/A")

        window=container.find("li",class_="DTBslk").find_next_sibling("li").find_next_sibling("li")
        windows_data.append(window.get_text(strip=True) if window else "N/A")

        storage=container.find("li",class_="DTBslk").find_next_sibling("li").find_next_sibling("li").find_next_sibling("li")
        storage_data.append(storage.get_text(strip=True) if storage else "N/A")

        display=container.find("li",class_="DTBslk").find_next_sibling("li").find_next_sibling("li").find_next_sibling("li").find_next_sibling("li")
        display_data.append(display.get_text(strip=True) if display else "N/A")

        price=container.find("div",class_="hZ3P6w DeU9vF")
        price_data.append(price.get_text(strip=True) if price else "N/A")
```

# A harampae of flipkart website
- Flipkart often limits search results to 40 or 42 pages for a single query. even if it shows 99 pages for laptops that i searched why
- - As the page number gets higher, the query gets slower and slower. To keep their website fast for everyone, they simply "cut off" the results after a certain point (like page 40).
- - Marketing teams want you to feel that Flipkart has "everything." By showing a total count of 30,000+ laptops, it builds trust and authority.

They know 95% of shoppers find what they want in the first 3 pages.

The remaining 5% will usually use a filter (Brand, RAM, Price) rather than clicking to page 80.

By the time you reach page 50, you aren't a "shopper" anymore; you are likely a competitor or a data scraper. By serving empty pages or redirects, they protect their data from being easily downloaded in bulk.

```python
df_dict={
    "Description":descriptions,
    "Processor":processor_data,
    "RAM":ram_data,
    "Window":windows_data,
    "Storage":storage_data,
    "Display":display_data,
    "Rating":rating,
    "Price":price_data
}

df=pd.DataFrame(df_dict)
laptop_data=df.to_excel("Laptops_data.xlsx",index_label="Laptop_Index")

# It shows only data of 41 pages so i could get only 984 rows
```
