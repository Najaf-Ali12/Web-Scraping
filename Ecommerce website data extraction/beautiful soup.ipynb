```python
import requests
```

```python
url="https://www.dawn.com/tech"
```

```python
def GetandSaveFiles(link,path):
    r=requests.get(link)
    with open(path,"w") as file:
        file.write(r.text)
```

```python
GetandSaveFiles(url, "scraped files/dawn.html")
```

```python
url2="https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"
r2=requests.get(url2)
print(r2)  # response[200] means html is received
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(r2.text,"lxml")  
print(soup)
```

requests.get(url) sends an HTTP GET request to the specified url. Its primary purpose is to retrieve the HTML content of that webpage (2:49-2:59).

BeautifulSoup(r.text, "lxml") then takes the raw HTML content obtained from the requests.get call (specifically r.text, which holds the HTML as text) and parses it using the lxml parser. This transforms the raw HTML string into a navigable Python object (the soup object) that allows you to easily search for and extract specific elements and data from the webpage's structure 

```python
# Extracting boxes
tag="div"
container_class="product-wrapper card-body"
boxes=soup.find_all(tag,class_=container_class)
print(len(boxes))
print(boxes)
```

```python
# extracting name of products
tag="a"
tag_class='title'
titles=soup.find_all(tag,class_=tag_class)
laptop_names=[]
for i in titles:
    laptop_names.append(i.get_text(strip=True).replace("..."," "))

print(len(laptop_names))
```

```python
# extracting prices
tag='h4'
price=soup.find_all(tag,class_="price float-end card-title pull-right")
prices=[]
for each in price:
    prices.append(each.get_text(strip=True).replace("$",""))

prices
```

```python
data={"laptop":laptop_names,"price":prices}
import pandas as pd
laptop_dataset=pd.DataFrame(data)
laptop_dataset.head()
```

```python
laptop_dataset.to_csv("scraped files\laptop_data.csv",index=False)
```
