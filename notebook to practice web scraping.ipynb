```python
import requests
```

```python
url="https://www.pcb.com.pk/"
r=requests.get(url)
print(r.status_code   )
print(r)  # this will tell status code in brackets
print(r.text)

# Status Codes and their meaning 
# Informational responses (100  199)
# Successful responses (200  299)
# Redirection messages (300  399)
# Client error responses (400  499)
# Server error responses (500  599)
# To get the meaning of different status codes visit the website given below
# https://developer.mozilla.org/en-US/docs/Web/HTTP/Reference/Status
```

```python
# Working with tags
from bs4 import BeautifulSoup
soup=BeautifulSoup(r.text,"lxml")
soup.div  # will print all the div container containing in soup html
soup.ul
print(soup.div.ul)
print("*"*100)
print(soup.div.li)
print(soup.li.a)
```

```python
# attributes examples
tag=soup.div  # soup.div is the shortcut of soup.find("div") , as it only keeps first tag so attributes of first tag are written
print(tag.attrs)  # to get the attributes in any specific tag
```

```python
# finding attributes of each div tag
tags=soup.find_all("div")
for each in tags:
    print(each)
    print(each.attrs)
    print("*"*100)
```

```python
print(soup.div.div.div.attrs)
print(soup.div.div.div.attrs.keys())  #  getting attribute of last div tag
print(soup.div.div.div.attrs["class"])  #  getting attribute of last div tag  
#soup.div.div.div.attrs is actually obtained as dictionary form so dictionary methods are applied
```

```python
# Getting the navigable string
tag=soup.div.button
print(tag.attrs)

tag2=soup.p
print(tag2.string)  # will print navigable string in p tag
```

```python
# Using find() method
desc=soup.find("div",{"class":"excerpt"})
print(desc.text)  # it will provide the text/navigable strings
print(desc.string) # this will return None and will not provide navigable strings
# because here the content is in another tag which is inside this div tag. There is no content directly in this div tag.
```

```python
time=soup.find("div",class_="meta")
print(time.text)
print(time.string)
```

# Difference between .text() and .string method
The core difference between the Beautiful Soup methods .text and .string lies in how they handle nested HTML tags:
- text retrieves all the text content within a tag and its descendants (children, grandchildren, etc.), combining them into a single string.
- string is much stricter. It only returns the text if the tag contains only one child and that child is a NavigableString (plain text) and not another tag [1, 2]. If the tag contains multiple children (text mixed with other tags, or multiple tags), it returns None


```python
# Getting data from multiple tags
mult_tags=soup.find_all(["p","li","nav"])
print(mult_tags)
```

```python
# find all filteration without Regular expressions
psl=soup.find_all(string="PCB")  # It will select just PCB
print(psl)
```

```python
# find_all() with Regular Expressions
import re
psl=soup.find_all(string=re.compile("PCB"))  # This will select every string that contains PCB
print(psl)
```

# selection using Attributes

```python
# Selecting the box with the attribute
attr_selection=soup.select('span[itemprop="price"]')  # the attribute must be in that tag not in another tag in that tag.
attr_selection[10].get_text(strip=True).replace("$","")
```

```python
# Selecting the box with the attribute
attr_selection=soup.select_one('span[itemprop="price"]')  # the attribute must be in that tag not in another tag in that tag.
attr_selection.get_text(strip=True).replace("$","")
```

# Tabular Data extraction

```python
# Selecting only CSK data from IPL auction website
url = "https://ticker.finology.in/"
r=requests.get(url)
soup=BeautifulSoup(r.text,"lxml")
print(soup)
```

```python
# extracting headers
headers=soup.select('th[scope="col"]')
for each in headers:
    print(each.get_text(strip=True))
    print("*"*200)
```
