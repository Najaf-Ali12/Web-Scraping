```python
dataset=[]
for url in urls:
    driver=webdriver.Chrome()
    driver.get(url)
    soup=BeautifulSoup(driver.page_source,"html.parser")
    com_cards=soup.find_all("div",class_="cos-cntr bg-grey")
    com_cards
    com_cards=soup.find_all("div",class_="co-list__itm")
    for card in com_cards:
    
        # checking for 2025 award
        award_tag_2025=card.find("div",class_="t-inner text-center")
        award_text_2025=award_tag_2025.get_text(strip=True) if award_tag_2025 else "No Award"
    
        # company name
        name_tag=card.find("p",class_="co-box__name font-medium")
        company_name=name_tag.get_text(strip=True).replace("Verified Company","") if name_tag else "Not mentioned"
        if not name_tag:   # This line of code will exclude the company whose name is not given on the website https://techbehemoths.com 
            continue
        
        # company location(city and country)
        location_tag=card.find("span",class_="co-box__loc-itm")
        company_location=location_tag.get_text(strip=True)if location_tag else "Location Not mentioned"
        company_city=company_location.split(",")[0]
    
        # company country
        company_country=company_location.split(",")[1]
        
        # getting source link
        source_tag=card.find("a")
        source="https://techbehemoths.com/"+source_tag.get("href") if source_tag else "Link not provided"
    
        # description
        desc_tag=card.find("p",class_="co-box__descr")
        description=desc_tag.get_text(strip=True) if desc_tag else "Description not provided"
    
        # team size
        team_size_tag=card.find("span",class_="value txt-blue")
        team_size=team_size_tag.get_text(strip=True) if team_size_tag else "Team size not mentioned"
    
        # hourly rate
        rate_tag=card.find("span",class_="co-box__tltip-txt flex-centered absolute")
        hourly_rate=rate_tag.get_text(strip=True) if rate_tag else "Hourly chargest not mentioned"
    
        # services
        service_tag=card.find("span",class_="txt")
        services=service_tag.get_text(strip=True) if service_tag else "Services not mentioned"
    
        # image
        image_tag=card.find("img")
        company_logo=image_tag.get("src") if image_tag else "Image not given"
    
        dataset.append({
            "Company":company_name,
            "City":company_city,
            "Country":company_country,
            "Award":award_text_2025,
            "Team Size":team_size,
            "Service":services,
            "Hourly Charges":hourly_rate,
            "Description":description,
            "Company logo":company_logo,
            "Link":source,
            "Data Source":"https://techbehemoths.com/"
        })


print(len(dataset))  # 3096
```
